## ðŸ’ Build 99% Confidence Interval Using Sample Standard Deviation

Assuming the sample is representative of the population, construct a 99% confidence interval for the mean number of characters printed before the print-head fails using the sample standard deviation. Explain the steps you take and the rationale behind using the t-distribution for this task.

```python
Mean=np.mean(Print_heads)         #.....Mean of sample
Std=np.std(Print_heads)           #.....Standard deviation of sample
N=len(Print_heads)                #.....Number of elements in sample

print('Mean of sample = ',Mean)
print('Standard deviation of sample = ',Std)
print('Number of elements in sample = ',N)
```

```python
# We have chosen the t-score as the population standard deviation is not known. 

Confidence_interval_percent=99
Degrees_freedom=N-1                #(Degrees_freedom=15-1=14)

t=2.977                            #The value of t obtained from t table corresponds to 99% and DF=14
```

```python
Upper_CI=round(Mean + t*(Std/np.sqrt(N)),3)
Lower_CI=round(Mean - t*(Std/np.sqrt(N)),3)

print('The 99% confidence limits are ',f'({Lower_CI},{Upper_CI})')
```

### â–¶The 99% confidence limits are  (1.095,1.382)
