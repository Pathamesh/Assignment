# ðŸ’ Association RulesðŸ’ 

```python
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data=pd.read_excel('Online retail.xlsx',header=None)
print("Sahpe of data ",data.shape)
data.head()
```

```python
data.isnull().sum()
```

## ðŸ’ Data pre-processing

```python
l=list(data[0])
L=[list(l[x].split(',')) for x in range(len(l))]
```

```python
TE=TransactionEncoder()
FD_array=TE.fit(L).transform(L)
FD=pd.DataFrame(FD_array,columns=TE.columns_)
print('Shape of FD ',FD.shape)
FD.head()
```

## ðŸ’ Using Apriori algorithm

```python
Bought_together=apriori(FD,min_support=0.009,use_colnames=True)
Bought_together
```

## ðŸ’ Mining association rules

```python
rules = association_rules(Bought_together, metric ="lift", min_threshold = 1) 
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules
```

## ðŸ’ Interview Questions:
#### 1.What is lift and why is it important in Association rules?
    The Lift metric is often used to measure the strength of association between two items. Lift is simply the ratio of the observed frequency of two items being bought together to the expected frequency of them being bought together if they were independent. In other words, Lift measures how much more likely two items will be bought together than would be expected if they were unrelated. A high Lift value indicates a strong association between two items, while a low Lift value indicates a weak association. The Apriori algorithm is designed to find itemsets with a high Lift value
    
#### 2.What is support and Confidence. How do you calculate them?
    The Support of an item is defined as the percentage of transactions in which an item appears.
    Item support can be calculated by finding the number of transactions containing a particular item divided by the total number of transactions
    â–¶ Support(X) = (Number of transactions containing X) / (Total number of transactions)
    
    Confidence is a statistical measure used in association rule learning that quantifies the likelihood that an association rule is correct. In other words, confidence measures the reliability of an association rule. The confidence of a rule is calculated as the ratio of the number of times the rule is found to be true to the total number of times it is checked.
    â–¶ Confidence(X -> Y) = Support(X âˆª Y) / Support(X)


#### 3.What are some limitations or challenges of Association rules mining?
    As datasets grow, the computational cost of finding frequent itemsets increases significantly, making it hard to apply on very large datasets.
    More variation in items increase dimensionality,complicating the mining process and leading to longer processing times.
    Rules may overlap or be similar making it difficult to choose most informative rule.
    The rules generated can be complex and difficult for non-experts to understand, limiting their practical use.
    Static association rules donâ€™t account for changes over time, which can lead to outdated or misleading insights.
