```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

##  ðŸ’ Data Cleaning and Preparation:

```python
data=pd.read_csv('Cardiotocographic.csv')
print('ðŸ’ Shape of data ',data.shape)
data.head()
```

```python
data.isnull().sum()
```

```python
data[data['LB'].isnull()]
```

```python
# Since all the 21 null values are in the same 21 records we can drop the null values 

data=data.dropna()
data.isnull().sum()
```

```python
data.info()                                               #Identify and correct any inconsistencies in data 
```

```python
plt.figure(figsize=(10,20))                        #Tendency[-1,0,1] and NSP[1,2,3] are categorical columns 
for y,x in enumerate(data.columns):
    v=data[x].value_counts()
    plt.subplot(7,3,y+1)
    plt.bar(v.index,height=v.values)
    plt.title(x)
plt.show()
    
```

```python
data=data[data['NSP'].isin([1,2,3])]                            #Cleaning the Categorical columns
data=data[data['Tendency'].isin([1,0,-1])]
data=data.reset_index().drop('index',axis=1)
data.head()
```

```python
plt.subplot(1,2,1)
plt.bar(data['Tendency'].value_counts().index,height=data['Tendency'].value_counts().values)
plt.title('Tendency')
plt.subplot(1,2,2)
plt.bar(data['NSP'].value_counts().index,height=data['NSP'].value_counts().values)
plt.title('NSP')

plt.show()
```

## ðŸ’ Outlier Treatment

```python
before_shape=data.shape
```

```python
Columns=['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV',
       'MLTV', 'Width']

plt.figure(figsize=(10,20))                         #Detect and treat outliers if necessary
for x,y in enumerate(Columns):
    plt.subplot(5,3,x+1)
    sns.boxplot(data[y])
    plt.title(y)
plt.show()
```

```python
Q1=data[Columns].quantile(0.25)
Q3=data[Columns].quantile(0.75)
IQR=Q3-Q1

Data = data[~((data[Columns] < (Q1 - 1.5 * IQR)) | (data[Columns] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print('Shape of data set before outlier treatment ',before_shape)
print('Shape of data set after outlier treatment ',Data.shape)
```

### ðŸš©Note: Removing outliers removes 50% of data so outlier treatment is not recommendable
